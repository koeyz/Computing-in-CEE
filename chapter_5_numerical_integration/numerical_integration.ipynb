## Numerical integration


### Mathematical definitions
Integration is the computation of an integral. For example, a function $f(x)$ can be integrated over an interval $x \in [a,b]$ as
 \begin{equation}
    I = \int_a^b f(x) dx
 \end{equation}
where $f(x)$ is called the integrand. Graphically, the integration of a function $f(x)$ over $[a,b]$ is equivalent of the "signed" area bounded by the curve $f(x)$, the $x$-axis, and the two vertical bounding lines. 

<img src="integration_area.png" width="500"/>
<h3 align="center">Figure. Integration over an interval.</h3> 

Integration can also be over a surface, which gives a volume.
 \begin{equation}
    I = \int_a^b \int_c^d f(x,y) dx dy
 \end{equation}
where $f(x,y)$ depends on both $x$ and $y$. Graphically, the integration of a function $f(x,y)$ over $[a,b] \times [c,d]$ is equivalent of the "signed" volume bounded by the curved surface $f(x,y)$, the $x-y$-plane, and the four vertical bounding surfaces. 

<img src="integration_volume.png" width="400"/>
<h3 align="center">Figure. Integration over a surface.</h3> 

Integration is the reverse of differentiation. If
 \begin{equation}
    F(x) = \int  f(x) dx
 \end{equation}
then
 \begin{equation}
    f(x) = \frac{d F(x)}{dx}
 \end{equation}
Here, $F(x)$ is called an antiderivative of $f(x)$. If $f(x)$ is a real-valued continuous function on a closed interval $[a,b]$, then 
  \begin{equation}
    I = \int_a^b f(x) dx = \left.F(x) \right\vert_{a}^{b} = F(b) - F(a)
 \end{equation}
 
Due to historical reasons, integration is also termed quadrature. Mathematically, there are two categories of integrals, i.e., definite and indefinite integrals in the forms of 
 \begin{equation}
 \int_a^b  f(x) dx    \quad \mathrm{vs.   } \int  f(x) dx
 \end{equation}
In this chapter, we are only interested in definite integrals.


### Application examples of numerical integration in CEE

In CEE, numerical integration is often used. The following is a short incomplete list of examples:

#### Example: Total discharge at a river cross-section
River discharge $Q$ at one particular cross-section is the volume of water flows passing the cross-section per unit time ($m^3/s$ or $ft^3/s$). To get $Q$, we need to know the flow velocity distribution across the cross section, which is usually obtained from measurements or computer simulations. The calculation of flow discharge is through the integration of velocity over the cross-sectional area.
\begin{equation}
Q = \int_S U ds
\end{equation}
where $U$ is flow velocity and $S$ is the river cross-section. Flowing water in rivers also carries dissolved and particulate matters, such as oxygen and sediment. It is often of interest to their fluxes too. To do that, we also need to know the concentration distribution $C$. With that, the flux is calcualted as another surface integration
\begin{equation}
Q = \int_S U C ds
\end{equation}

#### Example: Hydrostatic/hydrodynamic force on a structure or wind pressure force on a building 
To design structures like dams, bridges, or buidlings, it is important to calculate the load of water or wind on them. Regardless it is the hydrodstatic force on a dam or dynamic force on a bridge or wind load on a building, we need to know the pressure (normal) and shear stress (tangential) distribution on the surface of a structure.

Typically, the pressure contribution is dominating. Then, the pressure force can be calculated with the integration as follows:
\begin{equation}
\mathbf{F} = \int_\mathbf{S} p d\mathbf{s}
\end{equation}
where $p$ is pressure and $\mathbf{S}$ is area vector (magnitude = area, direction = normal).

#### Example: Length of a curved road (arc)
When we design a curved road or a highway, which is an arc, we need to calculate the total length. If we know the mathematical expression for the curve, we can calculate the length through integration as follows:
\begin{equation}
L = \int ds = \int \sqrt{\Delta x^2 + \Delta y^2} =\int \sqrt{1 + \left(\frac{\Delta y}{\Delta x}\right)^2} \Delta x = \int_a^b \sqrt{1+f'(x)^2} dx
\end{equation}
Here $L$ is the length of the arc and slope $f'(x) \approx \frac{\Delta y}{\Delta x}$.

#### Example: Center of mass
For structural and other analyses, the center of mass is an important property. For very simple shapes and density distributions, it is easy to calculate the center of mass either by hand or through a table. However, for more complicated geometry and density distribution, numerical integration is needed. 

For example, a beam is located between $x\in[a,b]$, with uniform cross-sectional shape, but nonuniform density $\rho(x)$. Its center of mass along the $x$-axis can be calculated as 
\begin{equation}
x_{\textrm{cm}} = \frac{\int_a^b x \rho(x) dx}{\int_a^b \rho(x)dx}
\end{equation}
where the numerator is the moment around $x$ = 0 and the denominator is the total mass of the beam.
 
<img src="center_of_mass_beam_nonuniform_density.png" width="400"/>
<h3 align="center">Figure. Center of amss for a beam with nonuniform density.</h3> 

Another example is a 2D plate with uniform density. The plate is formed by two curves $f(x)$ and $g(x)$ and the two curves intersect at $x = a$ and $x = b$. 

\begin{equation}
x_{\textrm{cm}} = \frac{M_y}{M} = \frac{\int_a^b x [f(x)-g(x)] dx}{\int_a^b [f(x)-g(x)]dx}
\end{equation}

\begin{equation}
y_{\textrm{cm}} = \frac{M_x}{M} = \frac{\int_a^b [f(x)^2-g(x)^2] dx}{2\int_a^b [f(x)-g(x)]dx}
\end{equation}

#### Example: Deflection of beams

Beams will deflect under loading. In any structural analysis textbook, there are tables for the deflection formulas if the loading regular, such as uniform or linear. For more complicated loading, the deflection needs to be solved with computer. In this example, a nonuniform load $w(x)$ is applied to a beam and induces nonuniform moment $M(x)$. The deflection $y$ is governed by the moment equation as
\begin{equation}
\frac{d^2y}{dx^2} = \frac{1}{EI(x)} M(x)
\end{equation}
where $E$ is the Elastic or Youngâ€™s modulus of the beam material, $I(x)$ is second Moment of Area of the beam's cross-section, and $M(x)$ is the bending moment at a section $x$ from the origin. Integrate the equation once will give the slope along the beam:
\begin{equation}
\frac{dy}{dx} = \int \frac{1}{EI(x)} M(x) dx + C_1
\end{equation}
Integrate the equation again will give the deflection along the beam:
\begin{equation}
  y(x) = \int \left[ \int \frac{1}{EI(x)} M(x) dx \right] dx + C_1x+C_2
\end{equation}
where $C_1$ and $C_2$ are integration coefficients depending on boundary conditions.

## Numerical integration
Numerical integration is the use of computer to calculate the approximation of an integration. The basic idea of numerical integration is to divide the integration interval into small pieces and then do summation. For example, for the following integration 
\begin{equation}
    I = \int_a^b f(x) dx
\end{equation}
the integration region $x\in[a,b]$ can be divived into many subregions, whose areas are approximated with different methods. All the different methods differ in how the subregion areas are calculated. It is also good to note that the integrand $f(x)$ can be in the form of tabulated data or a function.

<img src="numerical_integration_idea.png" width="300"/>
<h3 align="center">Figure. Idea of numerical integration.</h3> 


In this chapter, two categories of numerical integration methods are introduced. They are very popular in practice. 
- Newton-Cotes methods
- Gauss quadrature


### Newton-Cotes methods

In Newton-Cotes methods, the integrand is replaced with a simple polynomial which is easy to be integrated. It can be written as the following general equation:
 \begin{equation}
    I = \int_a^b f(x) dx \approx \int_a^b f_n(x) dx
 \end{equation}
 where 
 \begin{equation}
 f_n(x) =a_0 + a_1x + \ldots + a_{n-1}x^{n-1} + a_nx^n
 \end{equation}
 is a $n$-th order polynomial. The polynomial can be for the whole data points or as piecewise segments for each subregion. For example, the following figure shows that the integrant is approximated as a piecewise linear function.

<img src="newton_cotes_linear.png" width="300"/>
<h3 align="center">Figure. Newton-Cotes method where the integrant is approximated as a piecewise linear function.</h3> 

Three popular Newton-Cotes methods will be introduced here:
- trapezoidal rule
- Simpson's 1/3 rule
- Simpson's 3/8 rule

#### Trapezoidal rule
The trapezoidal rule is defined when the approximation polynomial is first order, i.e., the integrand curve on an interval is approximated as a linear line. The approximated area under the line is a trapezoid (thus the name). On an interval $[x_i,x_{i+1}]$, the integral (area) is
\begin{equation}
      I_i = (x_{i+1}-x_i)\frac{f(x_i)+f(x_{i+1})}{2}
\end{equation}

<img src="trapezoidal_rule.png" width="300"/>
<h3 align="center">Figure. Trapezoidal rule.</h3> 

If the integral interval $[a,b]$ is divided into $n$ segments of equal size $h$, the points on the $x$ axis can be denoted as $(x_0,x_1,\ldots,x_n)$. Then,
   \begin{equation}
      h = \frac{b-a}{n}
   \end{equation}
The total integral (area) is the summation of small areas, which can be expressed as
  \begin{eqnarray}
     I &=& h\frac{f(x_0)+f(x_1)}{2} + h\frac{f(x_1)+f(x_2)}{2} + \ldots + h\frac{f(x_{n-1})+f(x_n)}{2} \\
     & = & \frac{h}{2}\left[f(x_0) + 2 \sum_{i=1}^{n-1} f(x_i) + f(x_n) \right]
\end{eqnarray}     
   
<img src="trapezoidal_rule_all.png" width="300"/>
<h3 align="center">Figure. Trapezoidal rule with the summation of all small areas.</h3> 

To increase the accuracy of the integration with the Trapezoidal rule, you can increase the number of segments $n$. It is equivalent to reduce the segment size $h$. Indeed, the error of the Trapezodal rule follows
\begin{equation}
      E \propto -\frac{(b-a)^3}{12n^2}
\end{equation}
When $n$ doubles, the error is reduced to a quarter. This behavior is called \bverb|second| order convergence.

Next, let's do a manual calculatio example. Using the trapezoidal rule, calculate the following integration 
\begin{equation}
I = \int_0^1 (x^2+1) dx
\end{equation}
with 
- only one segment 
- two segments
- three segments
(Note the true solution is $I=\frac{4}{3}$)

The following is an implementation of the Trapezoidal rule. It is used the solve the same example integration as above.

```python
import numpy as np
import matplotlib.pyplot as plt
from math import exp  
import sys

#define the trapezoial rule: f is the integrand function
#a and b are the bound so of integration
#n is the number of segments 
def trapezoidal_rule(f, a, b, n):
    h = (b-a)/n
    s = (f(a) + f(b))
    for i in range(1,n):
        s = s + 2.0*f(a + i*h)
    
    return s*h/2.0

#the integrand function
def func(x):
    return x**2+1.0

a = 0
b = 1.0
n = 1000

I = trapezoidal_rule(func, a, b, n)

print("Trapezoidal rule integration result = ", I)
```

#### Simpson's rules
The trapezoidal rule uses a trapezoid to approximate the area under a segment. It uses a line connecting two points to approximate the integrant function. The idea of the Simpson's rules is to use higher-order polynomials to approximate the function. Two different Simpson's rules will be introduced. 
- on a segment with three points, a parabola can be fitted and this is called the 1/3 rule
- on a segment with four points, a third-order polynomial can be fitted and it is called the 3/8 rule

First, let's look at the Simpson's 1/3 rule. Let a segment be made of $[x_i,x_{i+2}]$ and $x_{i+1}$, with $x_{i+1}$ being  the mid-point. With the three points, a second-order Lagrange polynomial can be used to approximate the integrand
  \begin{eqnarray*}
  f_2(x) &=& \frac{(x-x_{i+1})(x-x_{i+2})}{(x_i-x_{i+1})(x_i-x_{i+2})} f(x_i) +  
             \frac{(x-x_{i})(x-x_{i+2})}{(x_{i+1}-x_{i})(x_{i+1}-x_{i+2})} f(x_{i+1}) \\
         & & + \frac{(x-x_{i})(x-x_{i+1})}{(x_{i+2}-x_{i})(x_{i+2}-x_{i+1})} f(x_{i+2})           
  \end{eqnarray*}
The integration is then
  \begin{eqnarray*}
    I &=& \int_{x_{i}}^{x_{i+2}} f(x)dx \approx \int_{x_{i}}^{x_{i+2}} f_2(x)dx \\
    &=& \int_{x_{i}}^{x_{i+2}} [ \frac{(x-x_{i+1})(x-x_{i+2})}{(x_i-x_{i+1})(x_i-x_{i+2})} f(x_i) +  
             \frac{(x-x_{i})(x-x_{i+2})}{(x_{i+1}-x_{i})(x_{i+1}-x_{i+2})} f(x_{i+1}) \\
         & & + \frac{(x-x_{i})(x-x_{i+1})}{(x_{i+2}-x_{i})(x_{i+2}-x_{i+1})} f(x_{i+2})    ] dx  
  \end{eqnarray*}

Let $h=(x_{i+2}-x_i)/2$, the integration gives
  \begin{eqnarray*}
  I &\approx&   (x_{i+2}-x_i) \frac{ \left[f(x_i) + 4f(x_{i+1}) + f(x_{i+2}) \right]}{6}  \\
    & = &  \frac{h}{{3}}\left[f(x_i) + 4f(x_{i+1}) + f(x_{i+2}) \right]
  \end{eqnarray*}
The name "Simpson's 1/3 rule" is from the "3" in the denominator.

If the integration interval $[a,b]$ is segmented with evenly distributed $n+1$ points where $n$ is an even number, there will be $n$ segments. Each segment has a length $h=(b-a)/n$. Then the total integration is 
  \begin{eqnarray*}
    I &\approx& 2h \frac{f(x_0)+4f(x_1)+f(x_2)}{6} + 2h \frac{f(x_2)+4f(x_3)+f(x_4)}{6} \\
      & &  + \ldots + 2h \frac{f(x_{n-2})+4f(x_{n-1})+f(x_n)}{6}
  \end{eqnarray*}
It can be simplified as
  \begin{equation}
  I \approx \frac{h}{3} \left[f(x_0) + 4 \sum_{i=1,3,5}^{n-1} f(x_i) + 2 \sum_{i=2,4,6}^{n-2} f(x_i) + f(x_n) \right]
  \end{equation}
An equivalent way to write the formula is
  \begin{equation}
  I \approx \frac{h}{3} \sum_{i=1}^{n/2}  \left[f(x_{2i-2}) + 4 f(x_{2i-1}) + f(x_{2i}) \right]
  \end{equation}
The error of the Simpson's 1/3 rule follows
   \begin{equation}
      E \propto -\frac{(b-a)^5}{180n^4}
   \end{equation}
which is fourth-order accurate because of the fact of $n^4$ in the denominator.  
  
Next, we show the use of the Simpson's 1/3 rule with a manual calculation example. Using the Simpson's 1/3 rule, calculate the following integration 
\begin{equation}
I = \int_0^1 (x^2+1) dx
\end{equation}
with 
- only one segment 
- two segments
(Note the true solution is $I=\frac{4}{3}$)  

Similar to Simpson's 1/3 rule, a third-order Lagrange polynomial can be used to approximate the integrand. In this case,  Four points are needed for a segment. Let a segment be $[x_i,x_{i+3}]$, and $x_{i+1}$ and $x_{i+2}$ are two mid-points.
Let $h=(x_{i+2}-x_i)/3$, similar analysis will give
  \begin{eqnarray*}
  I &\approx&  \frac{{3}h}{{8}}\left[f(x_i) + 3f(x_{i+1}) + 3f(x_{i+2}) + f(x_{i+3}) \right]
  \end{eqnarray*}
The name "Simpson's 3/8 rule" is from the "3/8" in the formula.

If the integration interval $[a,b]$ is segmented with evenly distributed $n+1$ points, there will be $n$ segments. The number $n$ needs to be $3m$ to use Simpson's 3/8 rule, where $m$ is a positive integer. With these, the following statements can be made:
- Each interval is $[x_{k-1},x_k]$ where $k$ =1, $3m$. 
- Each interval has a length of $h=\frac{b-a}{3m}$.
- The integration can be approximated as
  \begin{equation}
  I =\int_a^b f(x) dx \approx \frac{3h}{8} \sum_{k=1}^m \left[ f(x_{3k-3}) + 3 f(x_{3k-2}) + 3f(x_{3k-1}) + f(x_{3k}) \right]
  \end{equation}
  
It can be shown that the error of the Simpson's 3/8 rule is 
  \begin{eqnarray}
  E \propto - \frac{(b-a)^5}{80n^4}
  \end{eqnarray}
  
Manual calculation example: Using the Simpson's 3/8 rule, calculate the following integration 
\begin{equation}
I = \int_0^1 (x^2+1) dx
\end{equation}
with 
- only one segment 
- two segments
(Note the true solution is $I=\frac{4}{3}$)    

## Double integration

Double integral is integration over an area
\begin{equation}
  \int_c^d \left(\int_a^b f(x,y) dx \right) dy
\end{equation}   
which gives the (signed) volume between the $xy$ plane and the surface described by the function $f(x,y)$. The order of double integration is changeable because it will give same volume. 
\begin{equation}
  \int_c^d \left(\int_a^b f(x,y) dx \right) dy = \int_a^b \left(\int_c^d f(x,y) dy \right) dx
\end{equation}

<h3 align="center">Figure. Double integration over an area.</h3> 

To perform numerical double integration, it is not that different from one-dimensional integration. The double integration can be performed as follows: 
- integration in one-dimension with methods like trapezoidal or Simpson's rule with the values of the second dimension held constant,
- then integrate in the second dimension.
    
<img src="double_integral.png" width="300"/>
<h3 align="center">Figure. Numerical double integration.</h3>     

The following table shows the tabulated value at each grid point $(x_i, y_j)$, where $i\in [0,n]$ and $j \in [0,m]$. The rows and columns can be used to performed the integration in one-dimension and the other.

| :---:  |  :----:   |  :----:   |  :----:   |  :----:   |  :----:   |
|$y_0$ | $f(x_0,y_0)$ | $f(x_1,y_0)$  |  $f(x_2,y_0)$ | $\ldots$  |  $f(x_n,y_0)$  |
|$y_1$ | $f(x_0,y_1)$  | $f(x_1,y_1)$  |  $f(x_2,y_1)$ | $\ldots$  |  $f(x_n,y_1)$  |

Triple and multiple integrals are integration over a volume or higher dimensional space. The integration over one-dimension and followed by integration in other dimensions is no different than the doulbe integration.

The following is an implementation of double integration. Trapezoidal rule is used in both directions. The result is also compared with the result from using Python's "dblquad(...)" function.

```python
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt
from math import exp  
import sys

#define the trapezoial rule: f is the integrand values
#h is the segment size, and n is the number of segments 
def trapezoidal_rule(f, h, n):
    s = (f[0] + f[n])
    for i in range(1,n):
        s = s + 2.0*f[i]
    
    return s*h/2.0

#define the integrand fucntion
def integrand_func(x, y): 
    return pow(pow(x, 2) + pow(y, 2), 0.5) 

#perform the double integration 
#The number of grid points in x and y are [0,n] and [0,m], respectively
#The lower and upper bound in x and y are [xl,xu] and [yl,yu]
def doubleIntegration(n, m, xl, xu, yl, yu): 
    # z stores the fucntion values at grid points 
    # g stores the integral the integration value corresponding to each y   
    z = np.zeros((m+1,n+1))
    g = np.zeros(m+1)
    
    xs = np.linspace(xl,xu,n+1)
    ys = np.linspace(yl,yu,m+1)
    
    print("xs = ", xs)
    print("ys = ", ys)

    # calculate the segment size in x and y 
    delta_x = (xu - xl)/n 
    delta_y = (yu - yl)/m 

    # populate the integrand value table 
    for j in range(0, m+1): 
        for i in range(0, n+1): 
            z[j][i] = integrand_func(xs[i],ys[j]) 
            
    #print("z = \n", z.round(2))        

    # integrate with y held constant at each y value 
    for j in range(0, m+1): 
        g[j] = trapezoidal_rule(z[j,:],delta_x,n)    

    #print("g = \n", g.reshape(m+1,1).round(2))
    
    # integrate in y direction with the values calculated above
    answer = trapezoidal_rule(g,delta_y,m)
    
    
    return answer 


# Application of the code above

# xl and xu are upper and lower limit of x integral 
# yl and yu are upper and lower limit of y integral 
# n and m are number of segements in x and y
xl, xu, n = -1, 1, 10
yl, yu, m = -1, 1, 10
    
print("result from this program = ", doubleIntegration(n, m, xl, xu, yl, yu)) 
    
print("result from dblquad(...) = ", (integrate.dblquad(integrand_func,xl,xu,yl,yu))[0])
    
```

### Gauss quadrature
Newton-Cotes methods introduced above use evenly distributed data points $[x_0,x_1,\ldots,x_i,\ldots]$. The Newton-Cotes integral is: 
- Trapezoidal rule on an interval $[x_i,x_{i+1}]$: 
   \begin{equation}
      I \approx (x_{i+1}-x_i)\frac{f(x_i)+f(x_{i+1})}{2}
   \end{equation}  
- Simpson's 1/3 rule on an interval $[x_i,x_{i+2}]$:
  \begin{eqnarray*}
  I &\approx&   (x_{i+2}-x_i) \frac{ \left[f(x_i) + 4f(x_{i+1}) + f(x_{i+2}) \right]}{6}  
  \end{eqnarray*}
- Simpson's 3/8 rule on an interval $[x_i,x_{i+3}]$::
    \begin{eqnarray*}
  I &\approx& (x_{i+3}-x_i) \frac{\left[f(x_i) + 3f(x_{i+1}) + 3f(x_{i+2}) + f(x_{i+3}) \right]}{8}
  \end{eqnarray*}
  
All these integrals are weighted-average of function values at evenly distributed points. As you recall, 1D integration is to calculate the area:
  \begin{eqnarray}
  \textrm{Area = width} \times \textrm{height}
  \end{eqnarray}

These evenly distributed data points may not represent the integrand well. We have seen this in the interpolation chapter on the discussion of overfitting/oscillation with evenly distributed data points. Unevenly distributed data points may be better to capture the general trend of the integrand function. One such method is called Gauss quadrature.

To come up with the Gauss quadrature methods, the task is to determine the data points and weighting coefficients to achieve maximum accuracy (or even exact solution). 
  \begin{equation}
  \int_a^b f(x) dx \approx \sum_{i=0}^n a_i f(x_i)
  \end{equation}
  using $n+1$ data points.
An $(n+1)$-point Gauss quadrature rule is constructed to give an exact integration for polynomials of degree upto $2n+1$ (=$2(n+1)-1$). In this task, there are $2n+2$ unknowns:
- $n+1$ data points $x_i$ in $[a,b]$
- $n+1$ weighting coefficients $a_i$

The $2n+2$ unknowns can be derived based on the requirement that the rule exactly integrates a polynomial of degree upto $2n+1$. Because of this reason, the Gauss quadrature rule is more accurate if the integrand $f(x)$ behaves close to a polynomial of degree upto $2n+1$.

For simplicity, we will consider the integration region to be $[a,b]=[-1,1]$. Generalization will be made later because some linear transformation can be used later for different integration region. We will derive the data points and weighting coefficients with an example $n$ = 1 (two data points):
\begin{equation}
  \int_{-1}^{1} f(x')dx' \approx a_0 f(x'_0) + a_1 f(x'_1)
\end{equation}
The polynomial function $f(x')$ has a degree of $2n+1$ = 3, i.e., a cubic polynomial:
\begin{equation}
  f(x') = b_0 + b_1 x' + b_2 x'^2 + b_3 x'^3
\end{equation}
So plug into the definition, we have
  \begin{eqnarray*}
     \int_{-1}^{1} f(x') dx' &=& \int_{-1}^{1} \left(b_0 + b_1 x' + b_2 x'^2 + b_3 x'^3 \right) dx' \\
     &=& a_0 \left(b_0 + b_1 {x'}_0 + b_2 {x'}_0^2 + b_3 {x'}_0^3 \right) \\
     & & + a_1 \left(b_0 + b_1 {x'}_1 + b_2 {x'}_1^2 + b_3 {x'}_1^3 \right)
  \end{eqnarray*}
Collect similar terms with $b_0$, $b_1$, $b_2$ and $b_3$, we have
  \begin{eqnarray*}
  b_0 \left(a_0 + a_1 -\int_{-1}^1 dx'\right) + b_1 \left(a_0 {x'}_0 + a_1 {x'}_1 -\int_{-1}^1 {x'} d{x'}\right) & & \\
  b_2 \left(a_0 {x'}_0^2 + a_1 {x'}_1^2 -\int_{-1}^1 {x'}^2 d{x'}\right) + b_3 \left(a_0 {x'}_0^3 + a_1 {x'}_1^3 -\int_{-1}^1 {x'}^3 d{x'}\right)  = 0
  \end{eqnarray*}
Since the polynomial $f(x')$ is arbitrary, i.e., the coefficients $b_0$, $b_1$, $b_2$, and $b_3$ are arbitrary too. Then, for the above equation to be zero, all the terms in the parentheses have to be zero. 
   \begin{eqnarray*}
   a_0 + a_1 -\int_{-1}^1 dx' = 0 \quad
   a_0 {x'}_0 + a_1 {x'}_1 -\int_{-1}^1 x' dx' &=&0 \\
   a_0 {x'}_0^2 + a_1 {x'}_1^2 -\int_{-1}^1 x'^2 dx' = 0 \quad
   a_0 {x'}_0^3 + a_1 {x'}_1^3 -\int_{-1}^1 x'^3 dx' &=& 0
   \end{eqnarray*}  

We have four unknowns ($a_0$, $a_1$, $x_0$ and $x_1$) with four equations. The solution is as follows:
  \begin{eqnarray*}
  a_0 = 1 \quad a_1 = 1 \quad {x'}_0 = -\frac{\sqrt{3}}{3} \quad {x'}_1 = \frac{\sqrt{3}}{3}
  \end{eqnarray*}
Therefore
    \begin{equation}
  \int_{-1}^{1} f(x')dx' \approx a_0 f(x'_0) + a_1 f(x'_1) = f(-\frac{\sqrt{3}}{3}) + f(\frac{\sqrt{3}}{3})
  \end{equation}
This is our 2-point Gauss quadrature rule! It can exactly integrate a polynomial of degree upto 3!

Manual calculation example: integrate $f(x')=x'^2$ on $[-1,1]$ with $n=1$

Manual calculation example: integrate $f(x')=x'^3+2x'^2+1$ on $[-1,1]$ with $n=1$

Manual calculation example: integrate $f(x')=\cos(x')$ on $[-1,1]$ with $n=1$

We derived the Gauss quadrature on a nominal integration interval $x' \in [-1,1]$. What about a general interval $[a,b]$?
The solution is a simple linear transformation or change of variables from $x'$ to $x$. Let $x=\frac{b-a}{2}x'+\frac{b+a}{2}$, it is easy to show that 
  \begin{equation}
  \textrm{when } x' = -1, \quad x = a,\quad \textrm{and when } x' = 1, \quad x =b 
  \end{equation}
With this linear transformation, we have
  \begin{equation}
  dx = \frac{b-a}{2} dx'
  \end{equation}
  then
  \begin{equation}
  \int_a^b f(x)dx = \int_{-1}^1 f\left(\frac{(b-a)x'+b+a}{2} \right) \frac{b-a}{2} dx'
  \end{equation}
  
Manual calculation example: integrate $f(x)=x^2$ on $[a,b]=[1,3]$ with $n=1$ (note: analytical solution = $\frac{26}{3}\approx 8.6667$)  

The previous derivation and examples use $n=1$ and we get:
    \begin{equation}
  \int_{-1}^{1} f(x')dx' \approx a_0 f(x'_0) + a_1 f(x'_1)
  \end{equation}
  with 
    \begin{eqnarray*}
  a_0 = 1 \quad a_1 = 1 \quad {x'}_0 = -\frac{\sqrt{3}}{3} \quad {x'}_1 = \frac{\sqrt{3}}{3}
  \end{eqnarray*}
This is exact for polynomials up to the degree of $2n+1=3$. Can we go for \bverb|higher-order| polynomials?
    \begin{equation}
  \int_a^b f(x') dx' \approx \sum_{i=0}^n a_i f(x'_i)
  \end{equation}
It turns out $x'_i$ are roots of the Legendre polynomials:
  \begin{eqnarray*}
  P_0 &=& 1 \\
  P_1 &=& x \\
  P_2 &=& \frac{3x^2-1}{2} \\
  P_3 &=& \frac{5x^3-3x}{2} \\
  \ldots &=& \ldots
  \end{eqnarray*}
and a recursive relationship can be used to get higher-order Legendre polynomials:
  \begin{equation}
  P_n(x) = \frac{2n-1}{n}x P_{n-1}(x) - \frac{n-1}{n} P_{n-2}(x), \quad n=2,3,4,\ldots
  \end{equation}
For a $n$-point Gauss quadrature, the data points should be at the roots of $P_{n}$. For example, our 2-point Gauss quadrature rule, the data points are at the roots of $P_2$. The following figure shows the Legendre polynomials. Note the roots of each polynomial labeled as scatter points.

Keep in mind, these points $x_i$ are for the nominal interval $[-1,1]$. For general interval $[a,b]$, a linear transformation is needed. Weights $a_i$ do not depend on the interval.

## Python libraries for numerical integration

SciPy provides an "integrate" module:
[scipy.integrate documentation](https://docs.scipy.org/doc/scipy/reference/tutorial/integrate.html). It provides two categories of numerical integrations:

* integrating functions given function values at points: It provides trapezoidal (**trapz**), Simpson's rule (**simps**), etc. The documentation of the function **simps** is not clear on what Simpson's rule it implements. From reading the source code, it seems that the Simpson's 1/3 rule is implemented if the data points are evenly distributed. For unevenly distribbuted data points, the formula is slightly different. Some weighted-average 1/3 rule between two adjacent segments is used. 

* integrating functions given function definitions: It provides single, double, triple, and multiple integrations (**quad**, **dblquad**, **tplquad**, etc.).

The following is an example of the first category of numerical integration functionality in Python. The integration data points are generated on the interval $[1,5]$ with $x$ spacing $dx$ =1. The values $y$ are from the function $x^2$. The analytical solution is $\frac{124}{3}\approx41.33333...$.

```python
import numpy as np
from scipy import integrate

#set up the x and y values with only five data points
x = [1,2,3,4,5]
y = [1,4,9,16,25]

print("trapz result = ", integrate.trapz(y,x))
#print("trapz result = ", np.trapz(y,x))   !this also works

#set up the x and y values with ten data points
x = [1,1.5,2,2.5,3,3.5,4,4.5,5]
y = [1,2.25,4,6.25,9,12.25,16,20.25,25]

print("trapz result with refined data points = ", integrate.trapz(y,x))

#set up the x and y values with a lot more data points
x = np.linspace(1,5,10000)
y = x**2
print("trapz result with very refined data points = ", integrate.trapz(y,x))

#similary, you can also use the Simpson's rule
print("simps result = ", integrate.simps(y,x))
```

The following is an example of the second category of numerical integration functionality in Python. It uses the **quad()** function of the **scipy.integrate** module. The quad function returns the integral results and an esitmate of the errror. 

```python
from scipy.integrate import quad

# function we want to integrate
def func(x):
    return x**2

a = 1.0
b = 5.0

# call quad to integrate 
res, err = quad(func, a, b)

print("The integration result from quad(...) function is %f (+-%g)"%(res, err))
```

### Gauss quadrature and Legendre polynomials

Python's SciPy integrate module implements the Gauss quadrature in its **quadrature** function. Indeed, it implements an adaptive version of the Gauss quadrature with a specified tolerance. If the tolerance is not specified, a default value (~$10^{-8}$) is used.

```python
from scipy import integrate

# function we want to integrate
def func(x):
    return x**2

a = 1.0
b = 5.0

# call quadrature to integrate 
res, err = integrate.quadrature(func, a, b)

print("The integration result from quadrature(...) function is %f (+-%g)"%(res, err))
```

We know the data points $x_i$ for the Gauss quadrature are the roots of the Legendre polynomials. In Python, you can examine the Legendre polynomials, their roots, and even the Gauss quadrature weights. The following is an example Python code to plot several Legendre polynomials and their corresponding roots. In **SciPy**, the **legendre** sub-module is in its **special** module.

```python
from scipy.special import legendre

import matplotlib.pyplot as plt
import numpy as np

#plt.rc('text', usetex=True)  #allow the use of Latex for math expressions and equations
plt.rc('font', family='serif') #specify the default font family to be "serif"

min = -1.0
max = 1.0
step = 0.05

#generate the Legendre polynomials of degree up to 3, 
#i.e., P0, P1, P2, and P3.
for n in range(4):
    Pn = legendre(n)
    x = np.arange(min,max+step,step)
    y = Pn(x)
    plt.plot(x, y, label='P%d'%n)

#plot the roots of each Legendra polynomial above (except P0
#which has no root)
plt.scatter([], [], marker = 'x', s = 50)

#for P1
x = np.polynomial.legendre.leggauss(1)[0]
plt.scatter(x, [0], marker = '^', s = 50) 

#for P2
x = np.polynomial.legendre.leggauss(2)[0]
plt.scatter(x, [0,0], marker = 'o', s = 50) 

#for P3
x = np.polynomial.legendre.leggauss(3)[0]
plt.scatter(x, [0,0,0], marker = 'd', s = 50) 
    
#plot the x-axis    
plt.plot([-1,1],[0,0],'k--')
    
plt.xlim(-1.0,1.0)
plt.ylim(-1.0,1.1)

#set x and y axes label and font size
plt.xlabel('x',fontsize=16)
plt.ylabel('P(x)',fontsize=16)

#show the ticks on both axes and set the font size 
plt.tick_params(axis='both', which='major', labelsize=12)

plt.legend(loc='lower right',fontsize=14,frameon=False)
plt.savefig("legendre_polynomials.png",dpi=300, bbox_inches='tight', pad_inches=0)

plt.show()
```
